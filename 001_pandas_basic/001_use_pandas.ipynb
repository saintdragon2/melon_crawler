# 지난 주에 가장 인기 있던 KPOP 아티스트는??

```python
import pandas as pd
```

## CSV 파일 읽기
- pd.read_csv()

- 멜론 차트가 1~50위 까지, 그리고 51~100위 까지 두개의 CSV 파일에 저장되어 있음

```python
# read csv

df_latest_top_50 = pd.read_csv('../output/2025_03_24__2025_03_30장르종합주간차트____01_50.txt', sep='|')
df_latest_top_50
```

```python
df_latest_top_100 = pd.read_csv('../output/2025_03_24__2025_03_30장르종합주간차트____50_100.txt', sep='|')
display(df_latest_top_100)
```

```python
# 두 개의 dataframe을 합치기
df_latest = pd.concat([df_latest_top_50, df_latest_top_100], ignore_index=True)
df_latest
```

```python
# index가 중복되어 있다.문제가 없다면 문제가 없지만, 마음에 들지 않는다면...?
df_latest.head(60)
```

```python
# reset_index를 하면 되지만, 그 결과를 저장해야 함.
# - df_latest에 다시 저장하던지
# - 아니면, inplace=True를 추가하면 됨

df_latest.reset_index(inplace=True)
df_latest
```

```python
# 쓸데없는 column인 index가 생겨버렸다. 삭제하는 방법은? --> drop()을 사용하면 된다.
# - drop()은 기본적으로 새로운 dataframe을 반환한다.
# inplace=True를 써야 한다. 
df_latest.drop(columns=['index'], inplace=True)
df_latest
```

```python
# ranking을 점수로 변환한 새 column을 만들어보자. 
## 1등은 100점, 2등은 99점, 3등은 97점.....
## y = 101 - x 
## x:= 순위
df_latest['point'] = 101 - df_latest['rank']
df_latest
```

# 이번 주에 가장 인기있는 가수는 누구일까? 
- 가장 인기 있는 가수는 1위를 한 가수인가? 
- 아니면, top 100에 가장 많은 곡을 올린 가수인가? 
- 99등과 100등을 동시에 올린 가수가, 1등을 한 가수에 비해 인기가 있는 것인가? 

# 데이터 분석을 하려면, 기준이 명확해야 함!
- 가장 높은 포인트를 획득한 가수는 누구인가? 

```python
df_latest_artist = df_latest[['artist_name', 'artist_id', 'point']] # 필요한 세개의 컬럼만 남기기기
df_latest_artist
```

```python
# groupby를 사용하여 artist_id와 artist_name으로 그룹화한 후, point를 합산하여 새로운 데이터프레임을 생성합니다.
df_point_by_artist = df_latest_artist.groupby(['artist_id', 'artist_name']).sum()
df_point_by_artist
```

```python
# Sort
## 가장 높은 point를 획득한 가수는 누구인가?
df_point_by_artist.sort_values(by='point', ascending=False, inplace=True)
df_point_by_artist
```

```python
# reset_index를 사용하여 인덱스를 초기화합니다.
df_point_by_artist.reset_index(inplace=True)
df_point_by_artist
```

```python
# 몇 곡을 top 100에 올려놓았는지 알 수가 없다. 
## count로 갯수를 셀 수 있다. 
df_cnt_song = df_latest[['artist_id', 'artist_name', 'song_id']].groupby(['artist_id', 'artist_name']).count().sort_values(by='song_id', ascending=False)
df_cnt_song
```

```python
# column 명을 바꾸고 싶어! --> rename
# song_id를 song_cnt로!

df_cnt_song.rename(columns={'song_id': 'song_cnt'}, inplace=True)
df_cnt_song
```

```python
df_cnt_song.reset_index(inplace=True)
df_cnt_song
```

# 노래 제목까지 표에 보여주세요. 

```python

# 조건을 충족시키는 row만 골라내려면: loc
display(df_latest.loc[df_latest['artist_id']==6984])

# 이 중에서 원하는 column만 골라내려면, column 명
df_latest.loc[df_latest['artist_id']==6984]['song_title']

# 이걸 list로 만드려면
list(df_latest.loc[df_latest['artist_id']==6984]['song_title'])
```

```python
# 함수로 만들고, apply로 적용하기 
def get_song_titles(artist_id):
    return ';'.join(list(df_latest.loc[df_latest['artist_id']==artist_id]['song_title']))

get_song_titles(6984)
```

## apply
- artist_id의 값을 `get_song_titles` 함수에 넣어서, 그 결과 값을 `song_titles` 에 넣는다. 

```python
df_point_by_artist['song_titles'] = df_point_by_artist['artist_id'].apply(lambda x: get_song_titles(x))
df_point_by_artist
```
