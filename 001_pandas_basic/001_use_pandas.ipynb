```python
import pandas as pd
```

```python
# read csv

df_latest_top_50 = pd.read_csv('../output/2021_09_13__2021_09_19장르종합주간차트____01_50.txt', sep='|')
df_latest_top_50
```

```python
df_latest_top_100 = pd.read_csv('../output/2021_09_13__2021_09_19장르종합주간차트____50_100.txt', sep='|')
display(df_latest_top_100)
```

```python
# 두 개의 dataframe을 합치기

df_latest = df_latest_top_50.append(df_latest_top_100)
```

```python
# index가 중복되어 있다.문제가 없다면 문제가 없지만, 마음에 들지 않는다면...?
df_latest.head(60)
```

```python
# reset_index를 하면 되지만, 그 결과를 저장해야 함.
# - df_latest에 다시 저장하던지
# - 아니면, inplace=True를 추가하면 됨

df_latest.reset_index(inplace=True)
df_latest
```

```python
# 쓸데없는 column인 index가 생겨버렸다. 삭제하는 방법은?
# inplace=True를 써야 한다. 
df_latest.drop(columns=['index'], inplace=True)
df_latest
```

```python
# ranking을 점수로 변환한 새 column을 만들어보자. 
## 1등은 100점, 2등은 99점, 3등은 97점.....
## y = 101 - x 
## x:= 순위
df_latest['point'] = 101 - df_latest['rank']
df_latest
```

# 이번 주에 가장 인기있는 가수는 누구일까? 
- 가장 인기 있는 가수는 1위를 한 가수인가? 
- 아니면, top 100에 가장 많은 곡을 올린 가수인가? 
- 99등과 100등을 동시에 올린 가수가, 1등을 한 가수에 비해 인기가 있는 것인가? 

# 데이터 분석을 하려면, 기준이 명확해야 함!
- 가장 높은 포인트를 획득한 가수는 누구인가? 

```python
df_latest_artist = df_latest[['artist_name', 'artist_id', 'point']]
df_point_by_artist = df_latest_artist.groupby('artist_id').sum()
df_point_by_artist
```

```python
# Sort
## 가장 높은 point를 획득한 가수는 누구인가?
df_point_by_artist.sort_values(by='point', ascending=False, inplace=True)
df_point_by_artist
```

```python
# 누가 누군지 모르겠다. 
## artist_name column을 붙여야 함.
## df_latest_artist에 id와 artist명이 있음
## df_point_by_artist의 artist_id에  df_latest_artist를 붙이면 된다. 


df_artist_list = df_latest_artist[['artist_name', 'artist_id']]
df_artist_list.sort_values(by='artist_name')
```

```python
## 중복된 행을 지우려면 drop_duplicates

df_artist_list = df_artist_list.drop_duplicates().copy()
df_artist_list
```

```python
# 두개 dataframe을 합치려면 merge
df_point_by_artist = df_point_by_artist.merge(df_artist_list, on='artist_id')
df_point_by_artist.head(50)
```

```python
# 몇 곡을 top 100에 올려놓았는지 알 수가 없다. 
## count로 갯수를 셀 수 있다. 
df_cnt_song = df_latest[['artist_id', 'song_id']].groupby('artist_id').count().sort_values(by='song_id', ascending=False)
df_cnt_song
```

```python
# column 명을 바꾸고 싶어! --> rename
# song_id를 song_cnt로!

df_cnt_song.rename(columns={'song_id': 'song_cnt'}, inplace=True)
df_cnt_song
```

```python
df_point_by_artist = df_point_by_artist.merge(df_cnt_song, on='artist_id')
df_point_by_artist.head(30)
```

```python
# song_cnt 순으로 나열해보면.
df_point_by_artist.sort_values(by='song_cnt', ascending=False, inplace=True)
df_point_by_artist.head(20)
```

```python
# 함수로 만들고, apply로 적용하기 
def get_song_titles(artist_id):
    return ';'.join(list(df_latest.loc[df_latest['artist_id']==artist_id]['song_title']))

get_song_titles(672375)
```

```python
df_point_by_artist['song_titles'] = df_point_by_artist['artist_id'].apply(lambda x: get_song_titles(x))
df_point_by_artist
```
