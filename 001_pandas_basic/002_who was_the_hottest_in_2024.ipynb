# 2020년에 가장 HOT 했던 가수는 누구인가?
## 복습할 내용 + 새로 배울 내용
- 수십개의 charts를 합치기: concat (신규)
- chart마다 날짜 정보 입력하기 (복습 + 응용)

```python
# glob: 폴더 안의 파일을 둘러볼 때 쓴다. 

from glob import glob
```

```python
# 2024이 들어간 차트만 골라줘. 
charts_in_2024 = list()
for g in glob('../output/*'):
    if '2024' in g:
        charts_in_2024.append(g)

print(len(charts_in_2024))
```

```python

import pandas as pd

# sep='|'은 CSV 파일을 읽을 때 구분자를 |로 설정하는 것.
# charts_in_2024[0]은 glob으로 찾은 파일 중 첫 번째 파일을 의미한다.
pd.read_csv(charts_in_2024[0], sep='|')
```

```python
# 파일명만 가져오고 싶다면?
import os

# os.path.basename()을 사용하면 된다.
filename = os.path.basename(charts_in_2024[0])
print(filename)
```

```python
# 시작일과 종료일을 가져오고 싶다면?
# os.path.splitext()를 사용하면 된다.
from_to_str = filename[:22]
print(from_to_str)
```

```python
from_to = from_to_str.split('__')
print(from_to)
```

```python
df = pd.read_csv(charts_in_2024[0], sep='|')
df['from'] = from_to[0]
df['to'] = from_to[1]
df
```

```python
chart_list = list()

for g in charts_in_2024:
    filename = os.path.basename(g)
    from_to = g.split('\\')[1][:22].split('__')
    print(from_to)

    _df = pd.read_csv(g, sep='|')
    _df['from'] = from_to[0]
    _df['to'] = from_to[1]

    chart_list.append(_df)

display(chart_list[-1])

```

```python
# 자 이제 합쳐보자. concat

df_2024 = pd.concat(chart_list, axis=0)
df_2024
```

# 지난 시간 영상 복습
## point 추가, 곡 수 추가, 곡 제목 추가

```python
# point 추가
df_2024['point'] = 101 - df_2024['rank']
df_2024
```

```python
df_2024_by_artist = df_2024[['artist_id', 'artist_name',  'point']].groupby(['artist_id', 'artist_name']).sum()
df_2024_by_artist.sort_values(by='point', ascending=False, inplace=True)
display(df_2024_by_artist)
```

```python
df_2024_by_artist.reset_index(inplace=True)
df_2024_by_artist
```

# 아티스트별 곡 정리하기

```python
df_2024_artist_and_songs = df_2024[['artist_id', 'artist_name', 'song_title']]
df_2024_artist_and_songs.sort_values(by='artist_id', ascending=True, inplace=True)
df_2024_artist_and_songs
```

- 각 가수별로 어떤 곡들을 2024년에 차트에 등장시켰는지봐야 하므로, 중복 제거

```python
df_2024_artist_and_songs = df_2024_artist_and_songs.drop_duplicates()
df_2024_artist_and_songs.reset_index(inplace=True, drop=True)
df_2024_artist_and_songs
```

- 각 가수 별로 몇 곡을 2024년에 차트에 올렸는지 카운트하기

```python
df_2024_song_cnt_by_artist = df_2024_artist_and_songs[['artist_id', 'artist_name', 'song_title']].groupby(['artist_id', 'artist_name']).count()
df_2024_song_cnt_by_artist.sort_values(by='song_title', ascending=False, inplace=True)
df_2024_song_cnt_by_artist.reset_index(inplace=True)
display(df_2024_song_cnt_by_artist.head(20))
```

- 각 가수별로 차트에 올린 곡들을 df_2024_song_cnt_by_artist 에 새로운 칼럼으로 추가하기

```python
# 각 가수별로 차트에 올린 곡들을 df_2024_song_cnt_by_artist 에 새로운 칼럼으로 추가하기
def get_song_titles(artist_id):
    return list(df_2024_artist_and_songs.loc[df_2024_artist_and_songs['artist_id']==artist_id].drop_duplicates()['song_title'])



df_2024_song_cnt_by_artist['songs'] = df_2024_song_cnt_by_artist['artist_id'].apply(get_song_titles)
df_2024_song_cnt_by_artist
```

```python
# point 와 df_2024_song_cnt_by_artist 를 합쳐보자.
df_2024_artist = pd.merge(df_2024_by_artist, df_2024_song_cnt_by_artist, on=['artist_id', 'artist_name'])
df_2024_artist
```

```python
df_2024_artist.to_excel('./output/kpop_2024.xlsx')
```
