## 복습할 내용 + 새로 배울 내용
- 수십개의 charts를 합치기: concat (신규)
- chart마다 날짜 정보 입력하기 (복습 + 응용)
- datetime 양식 변환하기 (신규)

```python
# glob: 폴더 안의 파일을 둘러볼 때 쓴다. 

from glob import glob
```

```python
# 2020이 들어간 차트만 골라줘. 
charts_in_2020 = list()
for g in glob('../output/*'):
    if '2020' in g:
        charts_in_2020.append(g)

print(len(charts_in_2020))
```

```python

import pandas as pd

pd.read_csv(charts_in_2020[0], sep='|')
```

```python
# 여러 차트를 합쳐야 하기 때문에 새로운 column을 두개 만든다. 집계 시작일, 종료일
from_to = charts_in_2020[0].split('\\')[1][:22].split('__')
print(from_to)
df = pd.read_csv(charts_in_2020[0], sep='|')
df['from'] = from_to[0]
df['to'] = from_to[1]
df
```

```python
chart_list = list()

for g in charts_in_2020:
    from_to = g.split('\\')[1][:22].split('__')
    _df = pd.read_csv(g, sep='|')
    _df['from'] = from_to[0]
    _df['to'] = from_to[1]

    chart_list.append(_df)

display(chart_list[-1])

```

```python
# 자 이제 합쳐보자. concat

df_2020 = pd.concat(chart_list, axis=0)
df_2020
```

```python
# 지난 시간 영상 복습
# point 추가, 곡 수 추가, 곡 제목 추가
```

```python
# point 추가
df_2020['point'] = 101 - df_2020['rank']
df_2020
```

```python
df_2020_by_artist = df_2020[['artist_id', 'point']].groupby('artist_id').sum()
df_2020_by_artist.sort_values(by='point', ascending=False, inplace=True)
display(df_2020_by_artist)
```

```python
df_2020_by_artist = df_2020_by_artist.merge(df_2020[['artist_id', 'artist_name']].drop_duplicates(), on='artist_id')
df_2020_by_artist
```

```python
df_2020_by_artist.tail(30)
```

```python
df_2020_by_artist = df_2020_by_artist.merge(df_2020[['song_id', 'artist_id']].drop_duplicates().groupby('artist_id').count(), on='artist_id')
df_2020_by_artist
```

```python
df_2020_by_artist.rename(columns={'song_id': 'song_cnt'}, inplace=True)
df_2020_by_artist
```

```python
# 함수로 만들고, apply로 적용하기 
def get_song_titles(artist_id):
    return ';'.join(list(df_2020[['artist_id', 'song_title', 'song_id']].loc[df_2020['artist_id']==artist_id].drop_duplicates()['song_title']))


get_song_titles(672375)
```

```python
df_2020_by_artist['song_titles'] = df_2020_by_artist['artist_id'].apply(lambda x: get_song_titles(x))
df_2020_by_artist
```

```python
df_2020_by_artist = df_2020_by_artist.merge(df_2020[['rank', 'artist_id']].groupby('artist_id').min(), on='artist_id')
df_2020_by_artist
```
